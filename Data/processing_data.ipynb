```python
import pandas as pd
import numpy as np
import json
from sklearn.preprocessing import LabelEncoder

df = pd.read_csv('student_depression_dataset.csv')

print("Initial shape:", df.shape)
df.head()
```

```python
if 'id' in df.columns:
    df = df.drop(columns=['id'])

df['Financial Stress'] = pd.to_numeric(df['Financial Stress'], errors='coerce')
df['Financial Stress'] = df['Financial Stress'].fillna(df['Financial Stress'].median())

df['Sleep Duration'] = df['Sleep Duration'].astype(str).str.replace("'", "")
df['Dietary Habits'] = df['Dietary Habits'].astype(str).str.replace("'", "")
```

```python
category_mappings = {}

binary_map = {'Yes': 1, 'No': 0}
df['Have you ever had suicidal thoughts ?'] = df['Have you ever had suicidal thoughts ?'].map(binary_map)
df['Family History of Mental Illness'] = df['Family History of Mental Illness'].map(binary_map)
category_mappings['Have you ever had suicidal thoughts ?'] = binary_map
category_mappings['Family History of Mental Illness'] = binary_map

sleep_map = {
    'Less than 5 hours': 0,
    '5-6 hours': 1,
    '7-8 hours': 2,
    'More than 8 hours': 3
}
diet_map = {
    'Unhealthy': 0,
    'Moderate': 1,
    'Healthy': 2
}

df['Sleep Duration'] = df['Sleep Duration'].map(sleep_map)
df['Dietary Habits'] = df['Dietary Habits'].map(diet_map)
category_mappings['Sleep Duration'] = sleep_map
category_mappings['Dietary Habits'] = diet_map


df = df.dropna()

le = LabelEncoder()
categorical_cols = ['Gender', 'City', 'Profession', 'Degree']

for col in categorical_cols:
    df[col] = le.fit_transform(df[col])
    mapping_dict = dict(zip(le.classes_.astype(str), range(len(le.classes_))))
    category_mappings[col] = mapping_dict
```

```python
with open('category_mappings.json', 'w') as f:
    json.dump(category_mappings, f, indent=4)

df.to_csv('processed_data.csv', index=False)

print("Data cleaned, encoded, and saved.")
print("\nFinal Data Types:")
print(df.dtypes)
```
